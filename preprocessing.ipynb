```python
import pandas as pd

# First, convert the "Kms Driven" column to strings, then remove commas
car_Col["Kms Driven"] = car_Col["Kms Driven"].astype(str).str.replace(",", "", regex=False)

# Convert the column to numeric type (float or int)
car_Col["Kms Driven"] = pd.to_numeric(car_Col["Kms Driven"], errors='coerce')

# Display the cleaned data
print(car_Col["Kms Driven"].head())
```

```python


# Ensure all values in the "Price" column are strings and clean the column by removing unwanted characters
car_Col["Price"] = car_Col["Price"].astype(str)  # Convert to string
car_Col["Price"] = car_Col["Price"].str.replace("₹", "", regex=False)  # Remove currency symbol
car_Col["Price"] = car_Col["Price"].str.replace(",", "", regex=False)  # Remove commas
car_Col["Price"] = car_Col["Price"].str.lower()  # Convert to lowercase

def price_convert(price):
    # Check if the price contains "lakh"
    if "lakh" in price:
        # Extract numeric part, convert to float, and multiply by 100000
        return float(price.replace("lakh", "").strip()) * 100000
    # Check if the price contains "crore"
    elif "crore" in price:
        # Extract numeric part, convert to float, and multiply by 10000000
        return float(price.replace("crore", "").strip()) * 10000000
    else:
        try:
            value = float(price.strip())  # Convert the cleaned price to float
            # Multiply by 100000 for all values, including decimals
            return value  # Convert to int to avoid float representation
        except ValueError:
            return None  # Handle unexpected non-numeric values

# Apply the conversion function to the "Price" column
car_Col["Price"] = car_Col["Price"].apply(price_convert)

# Check the result
print(car_Col["Price"].head())
```

```python
def seat_convert(seat):
    # Ensure the input is a string, convert to lowercase, and remove extra spaces
    seat = str(seat).lower().strip()  # Convert to string here

    # Check if "seat" or "seats" is in the string
    if "seat" in seat or "seats" in seat:
        seat = seat.replace("seats", "").replace("seat", "").strip()  # Remove "seat" or "seats"
    
    # Try to convert remaining part to float
    try:
        return int(seat)  # Convert to float
    except ValueError:
        # Handle unexpected non-numeric values
        return None

# Apply the conversion function to the "Seats" column
car_Col["Seats"] = car_Col["Seats"].apply(seat_convert)

# Check the result
print(car_Col["Seats"].head())
```

```python
import numpy as np

def Engine_displacement_convert(engine):
    # Check if the value is NaN (to handle NaN floats properly)
    if pd.isna(engine):
        return None  # Return None for NaN values
    
    if isinstance(engine, str):  # Check if the value is a string
        engine = engine.lower().strip()  # Convert to lowercase and remove extra spaces
        if "cc" in engine:
            engine = engine.replace("cc", "").strip()
        try:
            return int(engine)  # Convert remaining part to integer
        except ValueError:
            return None  # Handle non-numeric values
    elif isinstance(engine, (int, float)):  # If already numeric, return as int
        return int(engine)
    else:
        return None  # Return None for unexpected types

# Apply the conversion function to the "Engine Displacement" column
car_Col["Engine Displacement"] = car_Col["Engine Displacement"].apply(Engine_displacement_convert)

# Check the result
print(car_Col["Engine Displacement"].head())
```

```python
def Mileage_convert(mileage):
    mileage = mileage.lower().strip()  # Convert to lowercase and remove extra spaces
    
    if "kmpl" in mileage:
        mileage = mileage.replace("kmpl", "").strip()  # Remove 'kmpl'
        try:
            return float(mileage)  # Convert to float
        except ValueError:
            return None  # Handle invalid mileage values
    elif "km/kg" in mileage:
        mileage_value = mileage.replace("km/kg", "").strip()  # Remove 'km/kg'
        try:
            return float(mileage_value) * 1.33  # Convert km/kg to kmpl using conversion factor
        except ValueError:
            return None  # Handle invalid mileage values
    else:
        try:
            return float(mileage)  # Convert remaining values to float
        except ValueError:
            return None  # Handle unexpected non-numeric values

# Apply the conversion function to the "Mileage" column
car_Col["Mileage"] = car_Col["Mileage"].apply(Mileage_convert)

# Check the result
print(car_Col["Mileage"].head())
```

```python
import re
import pandas as pd

def convert_power(power):
    # Handle NaN values
    if pd.isna(power):
        return None
    
    # Ensure the value is a string before applying string operations
    if isinstance(power, str):
        # Remove anything after "at" (if present)
        power = power.split("at")[0].strip()

        # Extract numeric part and unit using regular expression
        match = re.match(r"([0-9.]+)\s*([a-zA-Z]+)", power)
        
        if match:
            number = float(match.group(1))  # Extracted number
            unit = match.group(2).lower()   # Extracted unit (converted to lowercase)

            # Perform conversions based on the unit
            if unit == "ps":
                return number * 0.98  # PS to bhp
            elif unit == "kw":
                return number * 1.34  # kW to bhp
            elif unit == "bhp":
                return number  # bhp remains unchanged
            else:
                return number  # If unit is unknown, return the number as is

    # If no match found, or the value is numeric, try converting it directly
    try:
        return float(power)  # Try converting it directly if it's a number
    except (ValueError, TypeError):
        return None  # Return None for invalid values

# Apply the conversion function to the "Max Power" column
car_Col["Max Power"] = car_Col["Max Power"].apply(convert_power)

# Check the result
print(car_Col["Max Power"].head())
```

```python
import re
import pandas as pd

def convert_Torque(torque):
    # Handle NaN values first
    if pd.isna(torque):
        return None
    
    # Ensure the value is a string before applying string operations
    if isinstance(torque, str):
        # Remove anything after "at" (if present)
        torque = torque.split("at")[0].strip()

        # Extract numeric part and unit using regular expression
        match = re.match(r"([0-9.]+)\s*([a-zA-Z]+)", torque)
        
        if match:
            number = float(match.group(1))  # Extracted number
            unit = match.group(2).lower()   # Extracted unit (converted to lowercase)

            # Perform conversions based on the unit
            if unit == "kgm":
                return number * 9.806  # Convert kgm to Nm
            elif unit == "nm":
                return number  # Nm (Newton meters) remain unchanged

    # If no match found or the value is numeric, try converting it directly
    try:
        return float(torque)  # Try converting it directly if it's a number
    except (ValueError, TypeError):
        return None  # Return None for invalid values

# Apply the conversion function to the "Torque" column
car_Col["Torque"] = car_Col["Torque"].apply(convert_Torque)

# Check the result
print(car_Col["Torque"].head())
```

```python
def length_convert(length):
    # Ensure the input is a string, lowercase it, and remove extra spaces
    length = str(length).replace(",","").lower().strip()
    
    # Check if the length has "mm"
    if "mm" in length:
        return float(length.replace("mm", "").strip())  # Remove "mm" and convert to float
    
    # Check if the length has "kg" (we'll remove these cases)
    elif "kg" in length:
        return None  # We can return None for "kg" cases to drop them or mark them as NaN
    
    # Try to convert numeric values without any unit
    else:
        try:
            return float(length)  # Convert remaining part to float
        except ValueError:
            return None  # Handle unexpected non-numeric values

# Apply the conversion function to the "Length" column
car_Col["Length"] = car_Col["Length"].apply(length_convert)

# Check the result
print(car_Col["Length"].head())


```

```python
# Gear box


import re

def convert_Gearbox(speed):
    # Ensure the input is a string, convert it to lowercase, and strip extra spaces
    speed = str(speed).lower().strip()

    # Dictionary to handle word-based numbers (if needed)
    word_to_num = {
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9,
        "ten": 10
    }

    # Regular expression to match numeric values (e.g., "5", "6", "7 Speed", etc.)
    match = re.search(r"\b([4-9]|10)\b", speed)  # Matches numbers between 4 and 10
    if match:
        return int(match.group(1))  # Return the numeric value directly

    # Check if the string contains word-based numbers (e.g., "five", "six", etc.)
    for word, num in word_to_num.items():
        if word in speed:
            return num

    # Return None for all other cases like "CVT", "Direct Drive", or unexpected values
    return None


# Apply the conversion function to the "Gear Box" column
car_Col["Gear Box"] = car_Col["Gear Box"].apply(convert_Gearbox)

# Check the result
print(car_Col)
```

```python
import numpy as np

def Cargovol_convert(volume):
    # Ensure the input is a string and remove extra spaces and dashes
    volume = str(volume).strip().lower().replace("-", " ")
    
    try:
        # Check if the input is just a number
        if volume.replace('.', '', 1).isdigit():  # Allow for decimal points
            return float(volume)  # Return the standalone number as float
        
        # Split the value into number and unit
        num, unit = volume.split()
        
        # Check if the unit is related to "litres" or "liter"
        if unit in ["litres", "liter", "litre", "liters"]:
            return float(num)  # Convert the number part to float
        else:
            return np.nan  # Return NaN for other units
    except ValueError:
        return np.nan  # Handle errors when the value can't be split

# Apply the conversion function to the "Cargo Volume" column
car_Col["Cargo Volume"] = car_Col["Cargo Volume"].apply(Cargovol_convert)

# Display the cleaned data
print(car_Col["Cargo Volume"].head())
```

```python
def null_value_imputer(df):
    # Handling null values
    for i in df.columns:
        if (df[i].isnull().sum() / len(df)) * 100 >= 80:
            df.drop(i, axis=1, inplace=True)
        elif df[i].dtypes in ["int64", "float64"]:
            if df[i].skew() < 0.4 and df[i].skew() > -0.4:
                df[i].fillna(df[i].mean(), inplace=True)
            else:
                df[i].fillna(df[i].median(), inplace=True)
        else:
            df[i].fillna(df[i].mode()[0], inplace=True)

    print("Null values after imputation:\n", df.isnull().sum())
```

```python
import pandas as pd

# Select the columns for which you want to treat outliers
selected_cols = [
    "Kms Driven", 
    "No. of Owners", 
    "Price", 
    "Engine Displacement", 
    "Mileage", 
    "Car Age", 
    "Length",  # Make sure the casing matches your DataFrame
    "Width",   # Make sure the casing matches your DataFrame
    "Height",  # Make sure the casing matches your DataFrame
    "Max Power", 
    "Torque", 
    "Cargo Volume"  # Make sure the casing matches your DataFrame
]  

for column_name in selected_cols:
    print(f"Handling outliers for: {column_name}")

    # Adjust quantiles
    Q1 = car_data[column_name].quantile(0.35)  
    Q3 = car_data[column_name].quantile(0.95)   

    # Calculate IQR
    IQR = Q3 - Q1

    # Determine new lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = car_data[(car_data[column_name] < lower_bound) | (car_data[column_name] > upper_bound)]

    print(f"Number of outliers after adjustment: {len(outliers)}")
    print(outliers)

    # Handle outliers (for example, replace with mean)
    mean_value = car_data[(car_data[column_name] >= lower_bound) & (car_data[column_name] <= upper_bound)][column_name].mean()
    car_data[column_name] = car_data[column_name].mask(car_data[column_name] < lower_bound, mean_value)
    car_data[column_name] = car_data[column_name].mask(car_data[column_name] > upper_bound, mean_value)

    # Check the cleaned data
    final_outliers = car_data[(car_data[column_name] < lower_bound) | (car_data[column_name] > upper_bound)]
    print(f"Final number of outliers in {column_name}: {len(final_outliers)}\n")
```

```python
# Step 2: Calculate statistics for the cleaned data
stats_data = pd.DataFrame()

# Loop through each numeric column and calculate statistics
for col in selected_cols:
    mean = car_data[col].mean()
    median = car_data[col].median()
    mode = car_data[col].mode()[0] if not car_data[col].mode().empty else None  # Handle cases where mode is empty
    std = car_data[col].std()
    var = car_data[col].var()
    
    # Store the statistics in a DataFrame
    stats_data[col] = [mean, median, mode, std, var]

# Set row labels for clarity
stats_data.index = ['mean', 'median', 'mode', 'std', 'var']

# Display the statistics DataFrame
print("Statistics DataFrame:")
print(stats_data)
```

```python
# Separate features and target variable
x = scaled_df.drop(['Price'], axis=1)
y = scaled_df['Price']

# Compute correlation of each feature with the target variable
corr_target = scaled_df.corr()['Price'].abs()

# Define a function to select features based on correlation with the target variable
def select_features_by_correlation(corr_series, threshold):
    return corr_series[corr_series > threshold].index.tolist()

# Set the correlation threshold
corr_threshold = 0.05

# Select features based on the correlation threshold (excluding 'price')
selected_features = [feature for feature in select_features_by_correlation(corr_target, corr_threshold) if feature != 'Price']

# Add 'price' to the list of selected features
selected_features.append('Price')

# Create a DataFrame with selected features
x_selected = scaled_df.loc[:, selected_features]
```
