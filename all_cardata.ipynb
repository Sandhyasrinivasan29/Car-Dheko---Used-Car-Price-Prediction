```python
import pandas as pd

# Load the Excel file into a DataFrame
df_banglore = pd.read_excel("C:/Users/sandh/Downloads/bangalore_cars.xlsx", engine='openpyxl')


# Display the DataFrame
print(df_banglore.head())

```

```python
import pandas as pd
import ast

# Function to flatten the dictionary with unique keys
def flatten_dict(d, parent_key='', separator='_'):
    items = []
    for key, value in d.items():
        new_key = f'{parent_key}{separator}{key}' if parent_key else key
        if isinstance(value, dict):
            items.extend(flatten_dict(value, new_key, separator=separator).items())
        elif isinstance(value, list):
            for i, item in enumerate(value):
                if isinstance(item, dict):
                    items.extend(flatten_dict(item, f'{new_key}_{i}', separator=separator).items())
                else:
                    items.append((f'{new_key}_{i}', item))
        else:
            items.append((new_key, value))
    return dict(items)

# Columns to process
columns_to_process = ['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs']

# Initialize an empty list to store the flattened data
flattened_data = []

# Iterate through the DataFrame and flatten the necessary columns
for index, row in df_banglore.iterrows():
    flattened_row = {}
    for column in columns_to_process:
        # Check if column value is not empty before processing
        if pd.notna(row[column]):
            # Convert the text data into a Python dictionary using ast.literal_eval
            unstd_data = ast.literal_eval(row[column])
            # Flatten dictionary and append column name to avoid overwriting
            flattened_row.update(flatten_dict(unstd_data, parent_key=column))
    flattened_data.append(flattened_row)

# Create a DataFrame from the flattened data
std_df = pd.DataFrame(flattened_data)

# Process 'car_links' column (if not a nested structure, just copy as is)
if 'car_links' in df_banglore.columns:
    std_df['car_links'] = df_banglore['car_links']

# Add the 'City' column
std_df['City'] = 'Bangalore'

# Save the merged DataFrame to a CSV file
output_data = "std_bangalore_cars.csv"
std_df.to_csv(output_data, index=False)

# Display the merged DataFrame
df1 = pd.read_csv(output_data)
```
